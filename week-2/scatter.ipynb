```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/insurance.csv")

df.info()
```

```python

colors = []

for smoker in df["smoker"]:
    if smoker == "yes":
        colors.append("red")
    else:
        colors.append("blue")

smoker_df = df[df["smoker"] == "yes"]
non_smoker_df = df[df["smoker"] == "no"]
        
plt.scatter(smoker_df["bmi"], smoker_df["charges"], c="red", label="Smoker")
plt.scatter(non_smoker_df["bmi"], non_smoker_df["charges"], c="blue", label="Non-smoker")
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.legend()

plt.show()
```

```python

sns.relplot(df, x="bmi", y="charges", hue="smoker", style="sex", size="age")
```
