```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("imdb_top_1000.csv")

df.info()
```

```python
df["Runtime_int"] = df["Runtime"].str[:-4]

df["Runtime_int"] = pd.to_numeric(df["Runtime_int"])

df["Runtime_int"]
```

```python
df["year"] = pd.to_datetime(df["Released_Year"], errors="coerce")
```

```python
sns.relplot(df, x="year", y="Runtime_int", kind="line")
```
